# Joint Plot

```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns

tips = sns.load_dataset("C:\\Users\\Ankit Gaur\\seaborn-data\\tips")
```

```python
tips.head()
```

```python
tips.dtypes
```

```python
tips['size'] =tips.size.astype('int32')
```

```python
tips.dtypes
```

```python
g = sns.jointplot(x="total_bill", y="tip", data=tips)
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.plot(range(10))
g = sns.jointplot("total_bill","tip", data=tips, kind='reg')

```

```python
g = sns.jointplot("total_bill","tip", data=tips, kind="hex")
```

```python
g = sns.jointplot("total_bill","tip",data=tips,kind="kde")
```

```python
from scipy.stats import spearmanr
g = sns.jointplot("size","total_bill", data=tips, stat_func=spearmanr)
```

```python
g = (sns.jointplot("total_bill","tip",data=tips).plot_joint(sns.kdeplot, n_levels=6))
```

```python
g =(sns.jointplot("total_bill","tip",data=tips)
    .plot_joint(sns.kdeplot, n_levels=6)
    .plot_marginals(sns.rugplot))
```

# Facetgrid

```python
import seaborn as sns; sns.set(style="ticks", color_codes=True)
tips = sns.load_dataset("C:\\Users\\Ankit Gaur\\seaborn-data\\tips")
```

```python
tips.head()
```

```python
g = sns.FacetGrid(tips, col="day", row="sex")
```

```python
g = sns.FacetGrid(tips, col="time", row="smoker")
g = g.map(sns.distplot, "total_bill")
```

```python
g = sns.FacetGrid(tips, col="time", row="smoker")
g = g.map(sns.distplot, "total_bill", color='r')
```

```python
g = sns.FacetGrid(tips, col="time", row="smoker")
g = g.map(sns.scatterplot,"total_bill","tip","size")
```

```python
sns.scatterplot(data=tips, x="total_bill", y="tip", size="size")
```

```python
g = sns.FacetGrid(tips, col="time", row="smoker")
g = g.map_dataframe(sns.scatterplot,"total_bill","tip",size="size")
```

```python
g = sns.FacetGrid(tips, col="smoker", col_order=["No","Yes"])
g = g.map(sns.distplot, "total_bill", color="m")
```

```python
g = sns.FacetGrid(tips, col="smoker",row="sex")
g = (g.map(sns.scatterplot, "total_bill","tip", color="g")
    .set_axis_labels("total_bill (US Dollars)", "Tip"))
```

```python
g = sns.FacetGrid(tips, col="size", col_wrap=3)
g = (g.map(sns.distplot, "tip",color="c", kde=False)
     .set_titles("{col_name} dinners"))
```

# Reg Plots

```python
import numpy as np
import pandas as pd

import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
tips = sns.load_dataset('C:\\Users\\Ankit Gaur\\seaborn-data\\tips')
```

```python
tips.head()
```

```python
sns.regplot(x='total_bill', y='tip', data=tips , color="red")
```

```python
mean = [2,5]
cov =  [[1.1, 0.4], [2.2,3]]
x_value , y_value = np.random.multivariate_normal(mean, cov,100).T

sns.regplot(x=x_value,y=y_value)
```

```python
sns.regplot(x=x_value,y=y_value, color='purple')
```

```python
series_x_value = pd.Series(x_value, name="Var-X")
series_y_value = pd.Series(y_value, name="Var-Y")
sns.regplot(x=series_x_value , y=series_y_value)
```
